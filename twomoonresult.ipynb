## Original

```python
import json
from matplotlib import pyplot as plt
import numpy as np
json_file = open('results/twomoon_predef_False_alpha_0.1_gamma_0.3_eps_0.1_cor_0.01.json', 'r')

data=json.load(json_file)
aorvalues = data['aorvalues']
x_axis=np.arange(0,100)
# want to plot aor values with different colors in x_axis 0, 10

fig, ax = plt.subplots(figsize=(10, 10))
plt.title('AOR values for Two Moon dataset')
plt.xlabel('feature number')
plt.ylabel('AOR value')

bars = ax.bar(np.arange(len(aorvalues)), aorvalues)

if [0,10] is not None:
    for i in [0,10]:
        bars[i].set_color('r')
```
